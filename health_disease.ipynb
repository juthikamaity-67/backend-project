```python
import numpy as np
import pandas as pd
import joblib
import pickle
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier  # For KNN
from sklearn.ensemble import RandomForestClassifier # For Comparison
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
data=pd.read_csv("heart_disease_final_project.csv")
df=pd.DataFrame(data)
df
```

```python
df.shape
```

```python
df.isnull().sum().sum()       #check null value
```

```python
a=df.isnull().sum()/df.shape[0]*100
a
```

```python
df.isnull().sum().sum()
```

```python
cat_col=df.select_dtypes(include=["object"]).columns.tolist()
cat_col
```

```python
num_col=df.select_dtypes(include=["int64","float64"]).columns.tolist()
num_col
```

```python
cat_clean=SimpleImputer(strategy="most_frequent")
num_clean=SimpleImputer(strategy="mean")  #It automatically fills missing numerical gaps with the column's average value
```

```python
df[cat_col]=cat_clean.fit_transform(df[cat_col])
df[num_col]=num_clean.fit_transform(df[num_col])
df
```

```python
le = LabelEncoder()
df['severity'] = le.fit_transform(df['severity'])  #transfer string to int
```

```python
df.isnull().sum().sum()
```

```python
x= df.drop(['Target', 'risk_percentage', 'severity'], axis=1)
x.shape
```

```python
y = df[['Target', 'risk_percentage', 'severity']]
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=41) #split
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
std=StandardScaler()  #DATA SCALING
```

```python
x_train_sc=std.fit_transform(x_train)
x_test_sc=std.transform(x_test)
```

```python
print(x_train_sc.mean())
```

```python
# This will turn that long number into a clean 0.0
print(x_train_sc.mean().round(2))
```

```python
print("Std of Scaled Training Data:", x_train_sc.std())
```

```python
knn=KNeighborsClassifier(n_neighbors=5)     #DATA TRAIN
knn.fit(x_train_sc,y_train)
```

```python
y_pred = knn.predict(x_test_sc)  #  Make predictions on your scaled test data
y_pred
```

```python
y_probs = knn.predict_proba(x_test_sc)[0][:, 1] * 100     # Since knn is a multi-output classifier, predict_proba returns a list.
#display(y_probs)                                           # We access the first element [0] to get probabilities for the 'Target' column.
```

```python
target_accuracy = accuracy_score(y_test['Target'], y_pred[:, 0])   # Calculate accuracy for the 'Target' column (index 0)
print(f"Final Model Accuracy (Target): {target_accuracy * 100:.2f}%")

severity_accuracy = accuracy_score(y_test['severity'], y_pred[:, 2])  # Calculate accuracy for the 'severity' column (index 2)
print(f"Final Model Accuracy (Severity): {severity_accuracy * 100:.2f}%")
```
